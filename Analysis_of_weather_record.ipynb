## Data Analysis 

_This analysis is based on the weather_data folder obtained from the python crash book_

```python
from pathlib import Path
import matplotlib.pyplot as plt
import csv
from datetime import datetime

path = Path(r"C:/users/danhaya/arewads/arewads-python-project/weather_data/sitka_weather_2021_full.csv")
lines = path.read_text().splitlines()

reader = csv.reader(lines)
header_row = next(reader)

#extract average temperature and date
high_temp, dates, min_temp = [], [], []
for row in reader:
    current_date = datetime.strptime(row[2], "%Y-%m-%d")
    high = (row[7])
    mini = (row[8])
    dates.append(current_date)
    high_temp.append(high)
    min_temp.append(mini)

#Plot plot average temperature over time
plt.style.use("classic")
fig, ax = plt.subplots()
ax.plot(dates, high_temp, color = 'blue')
ax.plot(dates, min_temp, color= 'red')

#format plot
ax.set_title("Minimum and Maximum Daily Temperature, 2021", fontsize = 24)
ax.set_xlabel('', fontsize=16)
fig.autofmt_xdate()
ax.set_ylabel("Temperature (F)", fontsize=16)
ax.tick_params(labelsize=16)

plt.show

```

## visualization of Precipitation record

```python
from pathlib import Path
import csv
import matplotlib.pyplot as plt

path = Path(r"C:/users/danhaya/arewads/arewads-python-project/weather_data/sitka_weather_2021_full.csv")
lines = path.read_text().splitlines()

reader = csv.reader(lines)
header_row = next(reader)

#plot of precipitation over time
dates, ppts = [], []
for row in reader:
    current_date = datetime.strptime(row[2], '%Y-%m-%d')
    pft = row[5]
    #pt = int(row(5))
    dates.append(current_date)
    ppts.append(pft)

#plot precipitation distribution 
plt.style.use("bmh")
fig, ax = plt.subplots()
ax.plot(dates, ppts, color='green')

#format
ax.set_title("Precipitation Distribution")

plt.show()
```
